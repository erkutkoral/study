**3.1 Bavurular analiz ederek 0-18, 19-25, 26-40, 40+ ya aralklarna gre uu saylarn tablo ile memnuniyet dalmlarn ise (stun: satisfaction) grafik olarak gsterebilir misin?**

```python
max_age = df["Age"].max()
```

```python
max_age
```

```python
bins = [0, 18, 25, 40, max_age + 1]
```

```python
labels = ["0-18", "19-25", "26-40", "40+"]
```

```python
df["Age_Class"] = pd.cut(df["Age"],
                         bins = bins,
                         labels = labels)
```

```python
df["Age_Class"].value_counts()
```

```python
df["satisfaction"].value_counts()
```

```python
sns.countplot(x = 'Age_Class', hue = 'satisfaction', data = df, palette = 'viridis')

plt.title('Satisfaction Levels by Age Class')
plt.xlabel('Age Classes')
plt.ylabel('Count')

plt.legend(title='satisfaction')

plt.show()
```

**3.2 Type of Travel: Business travel olan ve 18 ya zerindeki mterileri filtreleyebilir misin? 
Veriyi analiz ettiinde bu mteri segmentinde satisfaction verisinin olumlu ya da olumsuz olmasna etki eden en byk faktrler nelerdir**?

```python
df["Type of Travel"].value_counts()
```

```python
sub_df = df[(df["Type of Travel"] == "Business travel") & (df["Age"] > 18)]
```

```python
sub_df.head()
```

```python
cat_cols = [col for col in df.columns if df[col].dtypes == "O"]
```

```python
num_cols = [col for col in df.columns if df[col].dtypes != "O"]
```

```python
num_cols = [col for col in num_cols if col != "id"]
```

```python
num_cols = [col for col in num_cols if col != "Age_Class"]
```

```python
cat_cols
```

```python
num_cols
```

```python
for col in num_cols:
    print(pd.DataFrame({col: sub_df.groupby("satisfaction")[col].mean()}),
         end = "\n\n\n")
```

```python
sub_df["satisfaction_encoded"] = sub_df["satisfaction"].map({"neutral or dissatisfied" : 0, "satisfied" : 1})
```

```python
cat_cols.append("satisfaction_encoded")
```

```python
for col in cat_cols:
    print(pd.DataFrame({"Target_Mean": sub_df.groupby(col)["satisfaction_encoded"].mean()}), end="\n\n\n")
```
